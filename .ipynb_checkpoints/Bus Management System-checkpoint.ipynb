```python
import tkinter as tk
import pyodbc
val_dict={}
class ScrolledFrame(tk.Frame):

    def __init__(self, parent, vertical=True, horizontal=False):
        super().__init__(parent)
        self._canvas = tk.Canvas(self)
        self._canvas.grid(row=0, column=0, sticky='news')  # changed
        self._vertical_bar = tk.Scrollbar(self, orient='vertical', command=self._canvas.yview)
        if vertical:
            self._vertical_bar.grid(row=0, column=1, sticky='ns')
        self._canvas.configure(yscrollcommand=self._vertical_bar.set)
        self._horizontal_bar = tk.Scrollbar(self, orient='horizontal', command=self._canvas.xview)
        if horizontal:
            self._horizontal_bar.grid(row=1, column=0, sticky='we')
        self._canvas.configure(xscrollcommand=self._horizontal_bar.set)
        self._vertical_bar.config(command=self._canvas.yview)
        self.inner = tk.Frame(self._canvas)
        self._window = self._canvas.create_window((0, 0), window=self.inner, anchor='nw')
        self.columnconfigure(0, weight=1)  # changed
        self.rowconfigure(0, weight=1)  # changed
        self.inner.bind('<Configure>', self.resize)
        self._canvas.bind('<Configure>', self.frame_width)

    def frame_width(self, event):
        canvas_width = event.width
        self._canvas.itemconfig(self._window, width=canvas_width)

    def resize(self, event=None):
        self._canvas.configure(scrollregion=self._canvas.bbox('all'))


class Input_Getter:
    def __init__(self, parent, text,id):
        self.parent = parent
        self.question = text
        self.create_widgets()
        self.id=id

    def get_input(self):
        value = self.entry.get()
        print('input id ',self.id+1, 'value:', value)
        
        val_dict['value '+str(self.id+1)]=value
    def create_widgets(self):
        self.labelframe = tk.LabelFrame(self.parent, text="")
        self.labelframe.pack(fill="both", expand=True)
        self.label = tk.Label(self.labelframe, text=self.question)
        self.label.pack(expand=True, fill='both')
        self.entry = tk.Entry(self.labelframe)
        self.entry.pack()
        self.entry.bind("<Return>", lambda x: self.get_input())
        

        

root = tk.Tk()
root.title("Taking Input from user")
root.geometry("400x300")
window = ScrolledFrame(root)
window.pack(expand=True, fill='both')
button = tk.Button(root, 
                   text="Quit and store to Database", 
                   fg="red",
                   command=root.destroy)
button.pack()



label=tk.Label(window.inner, text="How many values you want to enter?")
label.pack()
entry = tk.Entry(window.inner)
entry.pack()
def get():
    n_values = entry.get()
    # initializing an empty dictionary
    for i in range(int(n_values)):
        val_dict['value '+str(i+1)]=None
    for i in range(int(n_values)):
        Input_Getter(window.inner, "Enter Value "+ str(i+1),i)
    if int(n_values)>0:
        button1['state']=tk.DISABLED
    return n_values 

'''
button1 = tk.Button(window.inner, 
                   text="Click to submit",command=get)

button1.pack()


root.mainloop()

# creating a ms sql connection 
conn = ("Driver={SQL Server Native Client 11.0};"
            "Server=DESKTOP-2RD8KPT;"
            "Database=test;"
            "Trusted_Connection=yes;")
conn = pyodbc.connect(conn)

dbctrl=DataBaseController(conn)
dbctrl.delete() # Database previous values are deleted everytime this program runs 

# inserting to database
keys=val_dict.keys()
for key in keys:
    if val_dict[key] is not None:
        dbctrl.insert(key, val_dict[key] )
    else:
        dbctrl.insert(key, None )
dbctrl.read()
conn.close()
'''
```

```python
class Bus_Management:
    def __init__(self):
        self.tours=[]
        self.buses=[]
    def add_tour(self,pkg,buses,destination):
        self.tours.append(Tour(pkg,buses,destination))

class Tour:
    def __init__(self,pkg,buses,destination):
        self.package_name=pkg
        self.buses=buses
        self.destination=""
    def __repr__(self):
        return repr(self.package_name)
class Bus:
    def __init__(self,veh_id,cap):
        self.vehicle_no=veh_id
        self.capacity=cap
        
class Ticket:
    def __init__(self):
        self.user_name=''
        self.package=Tour()
        self.has_children=False
        self.no_passengers=0
        self.amount=0
        
```
